```python
import pandas as pd
import math
import matplotlib.pyplot as plt

planilha = pd.read_excel("/content/Planilha Aula.xlsx")
```

```python
#Quantidade de venda coluna
paises = planilha["País"]
contagem_pais = paises.value_counts()
print(contagem_pais)

segmento = planilha["Segmento"]
contagem_segmento = segmento.value_counts()
print(contagem_segmento)

categoria = planilha["Categoria"]
contagem_categoria = categoria.value_counts()
print(contagem_categoria)
```

```python
#Valor médio de vendas
receita = planilha["Vendas"]
print("receita média =",receita.mean())

#Lucro total
lucro = planilha["Lucro"]
print("lucro médio =",lucro.sum())
```

```python
#Vendas no brasil por estado

estado = planilha["Estado"]
paises = planilha["País"]
dados_brasil = planilha[planilha["País"] == "Brasil"]

estado_brasil = dados_brasil["Estado"]
vendas_brasil = dados_brasil["Vendas"]

plt.figure(figsize=(12, 6))

plt.bar(estado_brasil, vendas_brasil)

plt.xlabel('Estado')
plt.ylabel('Vendas')
plt.title('Vendas por Estado no Brasil')
plt.xticks(rotation=45, fontsize=8)

plt.subplots_adjust(bottom=0.2)

plt.show()
```

```python
#Lucro por segmento

vendas_por_segmento = planilha.groupby('Segmento')['Lucro'].mean()
vendas_por_segmento.plot(kind='bar')
plt.xlabel('Segmento')
plt.ylabel('Lucro')
plt.title('Lucro médio por Segmento')
plt.show()
```

```python
#Relações de Vendas

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

#relação entre Vendas e Lucro
axs[0].scatter(planilha['Vendas'], planilha['Lucro'])
axs[0].set_xlabel('Vendas')
axs[0].set_ylabel('Lucro')
axs[0].set_title('Correlação entre Vendas e Lucro')

#relação entre Desconto e Vendas
axs[1].scatter(planilha['Desconto'], planilha['Vendas'])
axs[1].set_xlabel('Desconto')
axs[1].set_ylabel('Vendas')
axs[1].set_title('Correlação entre Desconto e Vendas')

#relação entre Quantidade e Vendas
axs[2].scatter(planilha['Quantidade'], planilha['Vendas'])
axs[2].set_xlabel('Quantidade')
axs[2].set_ylabel('Vendas')
axs[2].set_title('Correlação entre Quantidade e Vendas')

plt.tight_layout()

plt.show()
```

```python
#Mais vendido
indice_produto_mais_vendido = planilha['Vendas'].idxmax()

valor_produto_mais_vendido = planilha.loc[indice_produto_mais_vendido]['Vendas']

produto_mais_vendido = planilha.loc[indice_produto_mais_vendido]['Nome do produto']

#Mais lucrativo
indice_produto_mais_lucrativo = planilha['Lucro'].idxmax()

valor_produto_mais_lucrativo = planilha.loc[indice_produto_mais_lucrativo]['Lucro']

produto_mais_lucrativo = planilha.loc[indice_produto_mais_lucrativo]['Nome do produto']

print("Produto mais vendido:", produto_mais_vendido, "...  Vendas =",valor_produto_mais_vendido )
print("Produto maior lucro:", produto_mais_lucrativo, "...  Lucro =",  valor_produto_mais_lucrativo)
```

```python
#Lucro total por região
lucro_por_regiao = planilha.groupby('Região')['Lucro'].sum()

plt.pie(lucro_por_regiao, labels=lucro_por_regiao.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Lucro por Região')
plt.show()
```
